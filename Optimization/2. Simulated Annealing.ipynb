Visualize the acceptance probabilities in Simulated Annealing for different values of energy differences and temperature

```python
# Import the required packages

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create dataframe

temperature_list = np.array([1000, 500, 100, 50, 10, 5, 1])
delta_energy_list = np.array([0.1, 1, 10, 100])

df = pd.DataFrame(index = temperature_list, columns = delta_energy_list)

for temperature in temperature_list:
    for delta_energy in delta_energy_list:
        value = np.exp(-delta_energy/temperature)
        df.loc[temperature, delta_energy] = value

df
```

```python
# Plot probability with temperature

for temperature in temperature_list:
    probability = np.exp(-delta_energy_list/temperature)
    
    plt.plot(
        delta_energy_list, probability,
        linestyle = '--', 
        label = 'Temp:{}'.format(temperature)
    )
    
    plt.xlabel('Energy Difference')
    plt.ylabel('Probability')
    plt.title('Probability variation with change in temperature')
    plt.legend()
```

```python
# Plot probability with delta_energy

for delta_energy in delta_energy_list:
    probability = np.exp(-delta_energy/temperature_list)
    
    plt.plot(
        temperature_list, probability,
        linestyle = '--', 
        label = 'DeltaE:{}'.format(delta_energy)
    )
    
    plt.xlabel('Temperature')
    plt.ylabel('Probability')
    plt.title('Probability variation with change in Energy Difference')
    plt.legend()
```
